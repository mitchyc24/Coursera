```python
import urllib.request
from bs4 import BeautifulSoup

site = "https://www.spacex.com/"

html = urllib.request.urlopen(site)
soup = BeautifulSoup(html, 'html.parser')
print(soup)

links = []
tags = soup('a')
for tag in tags:
    link = tag.get("href", None)
    links.append(link)
```
